```python
import numpy as np
df = pd.read_csv(io.StringIO(uploaded['도로살얼음 데이터정리1.csv'].decode('cp949')))
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
train_set, test_set = train_test_split(df, test_size=0.3, random_state=1006)

print(len(train_set), len(test_set))
```

```python
target = df.노면상태
```

```python
df = df.drop('노면상태', axis = 1)
```

```python
train_set = train_set.drop('노면상태', axis = 1)
```

```python
train_set
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3, random_state=1)
```

```python
X_test
```

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
```

```python
print(y_pred)
print(list(y_test))
```

```python
print('정확도 :', metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
rf_clf1 = RandomForestClassifier(n_estimators = 100, 
                                max_depth = 10,
                                min_samples_leaf = 8,
                                min_samples_split = 8,
                                random_state = 0,
                                n_jobs = -1)
rf_clf1.fit(X_train, y_train)
pred = rf_clf1.predict(X_test)
print('예측 정확도: {:.4f}'.format(accuracy_score(y_test,pred)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ftr_importances_values = rf_clf1.feature_importances_
ftr_importances = pd.Series(ftr_importances_values, index = X_train.columns)
ftr_top20 = ftr_importances.sort_values(ascending=False)[:20]

plt.figure(figsize=(9,7))
plt.title('Top 20 Feature Importances')
sns.barplot(x=ftr_top20, y=ftr_top20.index)
plt.show()
```
